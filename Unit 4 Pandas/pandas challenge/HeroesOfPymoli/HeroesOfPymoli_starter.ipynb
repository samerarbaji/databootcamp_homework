## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#print(purchase_data["Age"].max())
#print(purchase_data["Age"].min())

bins= (5,9,14,19,24,29,34,39,44)
binlabels=("<10", "10-14", "15-19", "20-24","24-29","30-34", "35-39", "40+")

purchase_data_df= pd.DataFrame(purchase_data)

purchase_data_df["Age Bracket"]= pd.cut(purchase_data_df["Age"],bins, labels= binlabels)


age_count = purchase_data_df.groupby('Age Bracket')['SN'].nunique()
age_count_df = pd.DataFrame(age_count)

age_count_df=age_count_df.rename(columns={'SN':'Total Count'})

age_count_df["Percentage of Players"]= round(age_count_df ['Total Count']/576*100,2)
age_count_df["Percentage of Players"]= age_count_df["Percentage of Players"].map("{:.2f}%".format)
age_count_df



```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchasecount= purchase_data_df.groupby("Age Bracket")["Purchase ID"].count()
purchase_count2_df= pd.DataFrame(purchasecount)

purchase_count2_df["Average Purchase Price"]= purchase_data_df.groupby("Age Bracket")["Price"].mean()
purchase_count2_df["Average Purchase Price"]= purchase_count2_df["Average Purchase Price"].map("${:.2f}".format)
purchase_count2_df["Total Purchase Value"]=purchase_data_df.groupby("Age Bracket")["Price"].sum()
purchase_count2_df["Total Purchase Value"]= purchase_count2_df["Total Purchase Value"].map("${:.2f}".format)

purchase_count2_df=purchase_count2_df.rename(columns={'Purchase ID':'Purchase Count'})

purchase_count2_df
#purchase_data_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
totalvalue = pd.DataFrame({"Total Purchase Value":purchase_data.groupby(["Item Name"])["Price"].sum()}).reset_index()

purchasecount = pd.DataFrame({"Purchase Count":purchase_data.groupby(["Item Name"])["Price"].count()}).reset_index()

itemprice = pd.DataFrame({"Item Price":purchase_data.groupby(["Item Name"])["Price"].mean()}).reset_index()


popular_items = totalvalue.merge(purchasecount, on="Item Name", how="left")

popular_items = popular_items.merge(itemprice, on="Item Name", how="left")
popular_items = popular_items.set_index('Item Name')

#popular_items["Item ID"]=purchase_data.groupby("Item Name")["Item ID"]

popular_items["Item Price"]= popular_items["Item Price"].map("${:.2f}".format)
popular_items = popular_items.sort_values(["Purchase Count"], ascending=False)

popular_items.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
popular_items = popular_items.sort_values(["Total Purchase Value"], ascending=False)
popular_items.head()
```
